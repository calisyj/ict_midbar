```python
import os
import pandas as pd
```

```python
# specify the directory where the xlsx files are located
directory = 'C:/Users/user/Desktop/capstone/data/03_env'
```

```python
# create an empty list to store the data from each file
data_frames = []
```

```python
# loop through each file in the directory
for filename in os.listdir(directory):
    if filename.endswith('.xlsx'):
        filepath = os.path.join(directory, filename)
        df = pd.read_excel(filepath)
        df['name'] = filename[0:5] # add the name of the xlsx file as a new column
        data_frames.append(df)
```

```python
# concatenate all of the data frames into one
combined_df = pd.concat(data_frames)
combined_df
```

```python
# specify the output file name and path
output_file = 'temporary_environment_first.xlsx'
```

```python
# write the combined data to the output file
combined_df.to_excel(output_file, index=False)
```
