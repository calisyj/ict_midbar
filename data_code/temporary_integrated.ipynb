```python
import pandas as pd
```

```python
# Load the first dataframe
df1 = pd.read_excel('temporary_preprocessed_environment_final.xlsx')
```

```python
df1['root_distance'] = 'NaN'
df1['root_thickness'] = 'NaN'
```

```python
df1
```

```python
# Load the second dataframe
df2 = pd.read_excel('temporary_root_distance_thickness_final.xlsx')
```

```python
df2
```

```python
# concatenate the DataFrames on the specified keys
merged_df = pd.merge(df1, df2, on=['time','name'], how='left')
merged_df[['root_distance', 'root_thickness']] = merged_df[['root_distance_y', 'root_thickness_y']].fillna(merged_df[['root_distance_x', 'root_thickness_x']])
merged_df.drop(['root_distance_x', 'root_thickness_x','root_distance_y', 'root_thickness_y'], axis=1, inplace=True)
df1 = merged_df
```

```python
df1
```

```python
# write the result to an Excel file
df1.to_excel("temporary_integrated.xlsx")
```
