```python
import os
import os
import xml.etree.ElementTree as ET
import numpy as np
import cv2
```

```python
class Masking:
    def __init__(self, name, width, height, root_points, br_points, dst_path):
        self.name = name
        self.width = width
        self.height = height
        self.root_points = root_points
        self.br_points = br_points
        self.dst_path = dst_path
    
    def polyline(self):
        img = np.zeros((int(self.height), int(self.width), 3), np.uint8) # height, width, bpp(bits per pixel)
        
        for root_point in self.root_points:
            s_root_points = str(root_point).split(";") # points
            xy_coord = list() # point_list
            for s_root_point in s_root_points:
                x_coord = int(float((str(s_root_point).split(",")[0])))
                y_coord = int(float((str(s_root_point).split(",")[1])))
                #print("root x_coord = {}, y_coord = {}".format(x_coord, y_coord))
                xy_coord.append([x_coord, y_coord])
            root = np.array(xy_coord, dtype=np.int32)
            cv2.polylines(img, [root], False, (0,0,255), 10)

        for br_point in self.br_points:
            s_br_points = str(br_point).split(";")
            xy_coord = list()
            for s_br_point in s_br_points:
                x_coord = int(float((str(s_br_point).split(",")[0])))
                y_coord = int(float((str(s_br_point).split(",")[1])))
                #print("br x_coord = {}, y_coord = {}".format(x_coord, y_coord))
                xy_coord.append([x_coord, y_coord])
            br = np.array(xy_coord, dtype=np.int32)
            cv2.polylines(img, [br], False, (255,0,0), 3)
        
        cv2.imwrite(os.path.join(self.dst_path, self.name), img)
        cv2.destroyAllWindows()
        print("{}_masking file is created".format(self.name))
```

```python
def xml_info(xml_path, dst_path):
    for path, dir, files in os.walk(xml_path):
        for file in files:
            if os.path.splitext(file)[-1] == ".xml" and os.path.getsize(os.path.join(path, file)) != None:
                tree = ET.parse(os.path.join(path, file))
                root = tree.getroot()

                for image in root.iter("image"):
                    name = image.get("name")
                    width = image.get("width")
                    height = image.get("height")
                    root_points = list()
                    br_points = list()
                    for polyline in image.iter("polyline"):
                        if polyline.get("label") == "root":
                            root_points.append(polyline.get("points"))
                        if polyline.get("label") == "br":
                            br_points. append(polyline.get("points"))
                    image = Masking(name, width, height, root_points, br_points, dst_path)
                    image.polyline()
                    



```
